```python
import sys
sys.path.append('.')
import json


from chroma import ChromaVectorStore

# Put all structured data into the data folder
data_path = "./data/sdsu_cs_courses.json"

# Initialize the Chroma vector store, and gets/initializes the default collections
vector_store = ChromaVectorStore()

#if you want to reset the collection between runs, uncomment the following line
#vector_store.reset_all_collections()

# add the cs classes to the all classes collection

with open(data_path, 'r', encoding='utf-8') as f:
    courses = json.load(f)

# courses is expected to be a list of dicts (each course object)
for index, course in enumerate(courses):
    document_text = str(course)
    vector_store.add_document(
        document_text=document_text,
        collection_name="allClasses"
    )
```

```python
print(f"Length of courses added to allClasses collection: {vector_store.all_classes_collection.count()}")
```

```python
class_data = vector_store.query_similar_documents("code CS 296, code CS 296, code CS 296 Experimental Topics", 1, "allClasses")
print(class_data)
```
